# Data Preprocessing

```python
df_employee = data_employee.copy()
```

## Data Duplicates

```python
# if the result is 1, then there is no duplicates
len(df_employee.drop_duplicates()) / len(df_employee)
```

Tidak ada duplikat yang ditemukan, sehingga **tidak ada treatment yang dilakukan**

## Missing Value

```python
# check missing value
df_employee.isnull().sum()
```

## Handling Invalid Value or Type of Data

```python
# column object
column_categories = df_employee.columns[df_employee.dtypes == 'object']

for col in column_categories:
  print(f"---{col}---")
  display(df_employee[col].value_counts())
  print('')
```

Tidak ditemukan placeholder yang menandakan misisng value, seperti -1, 99, NA, dsb

```python
# column numerical
column_numeric = df_employee.select_dtypes(include=['int64', 'float64']).columns
column_numeric = column_numeric[1:]


for col in column_numeric:
    print(f'-----{col}------')
    display(df_employee[col].value_counts())
    print('')
```

```python
# mengubah kolom `emp_id`
df_employee['emp_id'] = df_employee['emp_id'].astype(str)
```
